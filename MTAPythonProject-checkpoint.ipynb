```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "MTA_Key_Performance_Indicators__2008-2021 (1).csv"
df = pd.read_csv(file_path)
# Display basic information
print("Shape of the data:", df.shape)
print("\nColumn names:\n", df.columns)
print("\nData types:\n")
print(df.dtypes)

# Check for missing values
print("\nMissing values in each column:\n")
print(df.isnull().sum())
```

```python
# 2. Data Cleaning and Preparation
# Keep only necessary columns
df_cleaned = df[['Agency Name', 'Indicator Name', 'Category', 'Period Year', 'Period Month', 'YTD Actual', 'Monthly Actual']]

# Drop rows with important missing values
df_cleaned = df_cleaned.dropna(subset=['Indicator Name', 'YTD Actual'])

# Create a proper 'Date' column
df_cleaned['Date'] = pd.to_datetime(df_cleaned['Period Year'].astype(str) + '-' + df_cleaned['Period Month'].astype(str), errors='coerce')

# Show first few rows of cleaned data
print("\nCleaned data preview:\n")
print(df_cleaned.head())
```

```python
# 3. Line Plot - Trend Over Time for One KPI
indicator = "Collisions with Injury Rate"
df_kpi = df_cleaned[df_cleaned['Indicator Name'] == indicator]

plt.figure(figsize=(10,5))
plt.plot(df_kpi['Date'], df_kpi['YTD Actual'], marker='o')
plt.title(f'Trend Over Time: {indicator}')
plt.xlabel('Date')
plt.ylabel('YTD Actual')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#4. Yearly Trend Line Plot - Average YTD Actual
yearly_trend = df_cleaned.groupby('Period Year')['YTD Actual'].mean()

plt.figure(figsize=(8,5))
yearly_trend.plot(kind='line', marker='o')
plt.title('Average YTD Actual per Year')
plt.xlabel('Year')
plt.ylabel('YTD Actual')
plt.grid(True)
plt.show()
```

```python
 # 5. Histogram - Distribution of a KPI
plt.figure(figsize=(8,5))
sns.histplot(df_kpi['YTD Actual'], kde=True)
plt.title(f'Distribution of YTD Actual: {indicator}')
plt.xlabel('YTD Actual')
plt.ylabel('Frequency')
plt.show()
```

```python
#6. Correlation Heatmap - Relationship between Variables
numeric_df = df_cleaned[['YTD Actual', 'Monthly Actual']].dropna()

correlation = numeric_df.corr()

plt.figure(figsize=(6,5))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# 7. Box Plot - Outlier Detection in KPI
plt.figure(figsize=(10,6))
sns.boxplot(data=df_kpi, x='Period Year', y='YTD Actual')
plt.title(f'Outlier Detection: {indicator}')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#8. Bar Plot - First vs Last Year Comparison
year_summary = df_kpi.groupby('Period Year')['YTD Actual'].mean().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(data=year_summary, x='Period Year', y='YTD Actual')
plt.title(f'{indicator} - Yearly Averages')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 9. Top and Bottom Performing Years
top_years = year_summary.sort_values(by='YTD Actual', ascending=False).head(5)
bottom_years = year_summary.sort_values(by='YTD Actual', ascending=True).head(5)

print("\nTop Performing Years:\n", top_years)
print("\nWorst Performing Years:\n", bottom_years)
```
